```python
from google.colab import files
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
data = files.upload()
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import metrics
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
#The dataframe is displayed
df.head()
```

```python
#The missing values have bee dealt with
df.drop(["Cabin","Ticket"], axis = 1, inplace = True)
df.dropna(subset = ["Embarked"], inplace = True)
df.replace(to_replace="male", value=1, inplace=True)
df.replace(to_replace="female", value=-1, inplace=True)
df.replace(to_replace="S", value=1, inplace=True)
df.replace(to_replace="C", value=2, inplace=True)
df.replace(to_replace="Q", value=3, inplace=True)
df.drop(["Name"], axis = 1, inplace = True)
df.drop(["PassengerId"], axis = 1, inplace = True)
mean = df["Age"].mean(skipna = True)
df.fillna(mean, inplace=True)
```

```python
#Creating training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(df.drop(["Survived"], axis = 1),df["Survived"], test_size = 0.2, random_state=7)
```

```python
#Using scikit learn for logistic regression model
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
log_reg = LogisticRegression(max_iter=10000, random_state=0, penalty = "l2")
log_reg.fit(X_train, y_train.values)
X_test = scaler.transform(X_test.values)
y_pred = log_reg.predict(X_test)
```

```python
#This is the classification report
print(classification_report(y_test.astype(int), y_pred.astype(int)))
```
